```python
import openpyxl
import requests
import os
import multiprocessing#多线程
```

```python
root=r'/2022年报pdf'#存放pdf年报的位置
start=4799#文件中的第几行开始
end=4798#文件中的第几行结束
```

```python
wb=openpyxl.load_workbook('2022all_url_data.xlsx')
ws=wb.active
```

```python
def down():
    for i in range(start,end+1):
        try:
            dir_name=ws.cell(i,1).value+'-'+ws.cell(i,2).value
            pdf_name=ws.cell(i,1).value+'-'+ws.cell(i,7).value+'年度报告'
            dir_path=root+'\\'+dir_name
            pdf_path=dir_path+'\\'+pdf_name+'.pdf'
            if not os.path.exists(dir_path):
               os.makedirs(dir_path)
            pdf_url=ws.cell(i,6).value
            response=requests.get(pdf_url)
            data=response.content
            with open(pdf_path,mode='wb') as f:
               f.write(data)
            print(f'正在下载：{pdf_path}')
        except:
            ws.cell(i,5).value='有错误，请检查'
    wb.save('all_url_data.xlsx')
```

```python
if __name__ == '__main__':
    dir_process = multiprocessing.Process(target=down())
    dir_process.start()
```
